# **펀드리스트 크롤링**

```
import requests
import xml.etree.ElementTree as ET
import pandas as pd

def get_fund_data(): # 모든 펀드 데이터 
  payload = """<?xml version="1.0" encoding="utf-8"?>
  <message>
    <proframeHeader>
      <pfmAppName>FS-DIS2</pfmAppName>
      <pfmSvcName>TpExpertMgrFundSrchSO</pfmSvcName>
      <pfmFnName>selectFund</pfmFnName>
    </proframeHeader>
    <systemHeader></systemHeader>
      <DISExpertManagerDTO>
      <val2></val2>
      <val5>N</val5>
      <val1></val1>
      <val3></val3>
      <val4></val4>
  </DISExpertManagerDTO>
  </message>
  """

  r = requests.post('http://dis.kofia.or.kr/proframeWeb/XMLSERVICES/', data=payload)

  # XML to DataFrame
  tree = ET.XML(r.content) 
  all_records = []

  element = tree.findall('message/DISExpertManagerListDTO/list')
  for i, child in enumerate(element):
      record = {}
      for i, subchild in enumerate(child):
          record[subchild.tag] = subchild.text
      all_records.append(record)

  df = pd.DataFrame(all_records)

  # 컬럼 번호 순서대로
  cols = ['val1', 'val2', 'val3', 'val4', 'val5', 
          'val6', 'val7', 'val8', 'val9', 'val10', 
          'val11', 'val12', 'val13', 'val14', 'val15', 
          'val16', 'val17', 'val18', 'val19', 'val20', 
          'val21', 'val22', 'val23', 'val24', 'val25', 
          'val26', 'val27', 'val28', 'val29', 'val30',
  ]
  df = df[cols]
  
  # 컬럼 rename
  column_map = {
    'val1':'펀드코드', 
    'val2':'운용사코드', 
    'val3':'운용사', 
    'val4':'펀드명', 
    'val5':'펀드매니저수', 
    'val6':'최근3개월변경_등록', 
    'val7':'최근3개월변경_말소', 
    'val8':'일자', 
    'val9':'기준일자',
    'val10':'설정원본', 
    'val11':'순자산총액', 
    'val12':'수익률_기준일자', 
    'val13':'수익률_1개월', 
    'val14':'수익률_6개월', 
    'val15':'수익률_1년', 
    'val16':'수익률_설정전환일',
    'val17':'보수등_기준일자', 
    'val18':'보수등_총보수', 
    'val19':'보수등_수수료율', 
    'val20':'보수등_TER', 
    'val21':'보수등_매매중계수수료율', 
    'val22':'val22',
    'val23':'계층',
    'val24':'val24',
    'val25':'val25',
    'val26':'val26',
    'val27':'val27',
    'val28':'val28',
    'val29':'모펀드코드', 
    'val30':'val30',
  }
  df.rename(columns=column_map, inplace=True)

  # 사용 컬럼 지정
  use_cols = ['펀드코드', '펀드명', '운용사코드', '운용사', '기준일자', '순자산총액', 
              '수익률_기준일자', '수익률_1개월', '수익률_6개월', '수익률_1년', '수익률_설정전환일'
  ]
  return df[use_cols]
```

```
import requests
import xml.etree.ElementTree as ET
import pandas as pd

def get_fund_manager_detail(manager_id, date): # 펀드 매니저 상세
  date = pd.to_datetime(date)
  
  payload_tmpl = """<?xml version="1.0" encoding="utf-8"?>
  <message>
    <proframeHeader>
      <pfmAppName>FS-DIS2</pfmAppName>
      <pfmSvcName>DISExpertMgrBaseInfoSO</pfmSvcName>
      <pfmFnName>selectFund</pfmFnName>
    </proframeHeader>
    <systemHeader></systemHeader>
      <DISExpertManagerDTO>
      <val1>{manager_id}</val1>
      <val2>{date}</val2>
  </DISExpertManagerDTO>
  </message>
  """
  if date: 
    payload = payload_tmpl.format(manager_id=manager_id, date=date.strftime('%Y%m%d'))
  else:
    payload = payload_tmpl.format(manager_id=manager_id, date="2019-09-02")
  
  r = requests.post('http://dis.kofia.or.kr/proframeWeb/XMLSERVICES/', data=payload)

  # XML to DataFrame
  tree = ET.XML(r.content) 
  all_records = []

  element = tree.findall('message/DISExpertManagerListDTO/list')
  for i, child in enumerate(element):
      record = {}
      for i, subchild in enumerate(child):
          record[subchild.tag] = subchild.text
      all_records.append(record)

  df = pd.DataFrame(all_records)
  if df.empty:
    return  None
  
  print(df)
  # 컬럼 번호 순서대로
  cols = ['val1', 'val2', 'val3', 'val4', 'val5', 
          'val6', 'val7', 'val8', 'val9', 'val10', 
          'val11', 'val12', 'val13', 'val14', 'val15', 
          'val16', 'val17', 'val18', 'val19', 'val20', 
          'val30']
  df = df[cols]
  
  # 컬럼 rename
  column_map = {
    'val1':'매니저코드', 
    'val2':'펀드코드', 
    'val3':'펀드명', 
    'val4':'운용시작일', 
    'val5':'운용종료일', 
    'val6':'설정원본', #백만원 
    'val7':'순자산총액', #백만원 
    'val8':'펀드유형', 
    'val9':'투자지역',
    'val10':'수익률_1개월', 
    'val11':'수익률_6개월', 
    'val12':'수익률_1년', 
    'val13':'수익률_설정이후', 
    'val14':'보수등_총보수', 
    'val15':'보수등_수수료율', 
    'val16':'보수등_TER',
    'val17':'보수등_매매중계수수료율', 
    'val18':'매니저구분', 
    'val19':'계층', 
    'val20':'설정일', 
    'val30':'모펀드코드'
  }
  df.rename(columns=column_map, inplace=True)
  
  # 컬럼 타입 변환
  float_cols = ['수익률_1개월', 
  '수익률_6개월', 
  '수익률_1년', 
  '수익률_설정이후', 
  '보수등_총보수',
  '보수등_수수료율', 
  '보수등_TER',
  '보수등_매매중계수수료율'
               ]
  
#  for col in float_cols:
#    for row in df[col]:
#      if type(row) == 'str':
#        df[float_cols].loc[row] = '0'
        
#  if df[float_cols].dtypes == 'str':
   # df[float_cols]= df.fillna('missing')


#      else:
  df[float_cols] = df[float_cols].apply( pd.to_numeric, errors='coerce' )

  df[float_cols] = df[float_cols].astype(float)
  
      

  # 사용 컬럼 지정
  use_cols = ['매니저코드', '펀드코드', '펀드명', '운용시작일','매니저구분','펀드유형', 
              '설정원본', '순자산총액', '투자지역', 
              '수익률_1개월', '수익률_6개월', '수익률_1년', '수익률_설정이후','보수등_총보수',
  '보수등_수수료율', '보수등_TER','보수등_매매중계수수료율'
  ]
  return df[use_cols]
```
